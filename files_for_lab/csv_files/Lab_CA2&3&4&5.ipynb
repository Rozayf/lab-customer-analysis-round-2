```python
def clean(x):
    if x in ['Yes']:
        return 1
    else:
        return 0

data_frame1['response rate'] = list(map(clean, data_frame1['response']))
data_frame1['response rate'].unique()
```

```python
data_frame1
```

```python
sns.histplot(bins=10, x="total claim amount", hue="response", data=data_frame1, stat="percent")
plt.show()
```

```python
def income_bins(x):
    if x <20000:
        return 1
    elif x<40000:
        return 2
    elif x<60000:
        return 3
    elif x<80000:
        return 4
    else:
        return 5
    
data_frame1['income_bins'] = list(map(income_bins, data_frame1['income']))

data_frame1
```

```python
data_frame1.groupby('income_bins')['response rate'].mean().plot.bar()
plt.show()
```
